> ### 3. Check whether all the zones generate the same amount of sales.

We can check whether all the zones generate the same amount of sales or not, through an ANOVA test. For this, two hypothesis are defined initially.
> **Null Hypothesis** : There is no difference in the sales values among the zones.

> **Alternate Hypothesis** : There is atleast one zone with sales values different from other zones.

```python
statistic, pvalue = f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print(f"The F-statistic value is : {statistic:.3f}")
print(f"The Calculated p value is : {pvalue:.3f}")
alpha = 0.05
if pvalue > alpha:
    print("Accept the Null Hypothesis")
elif pvalue < alpha:
    print ("Reject the Null Hypothesis")
```

As the 'pvalue' is less than the 'alpha' value, we will have to reject the hypothesis. Thus, it can be concluded from the test that there is atleast one zone with sales values different from other zones. All the zones generate different amount of sales.
